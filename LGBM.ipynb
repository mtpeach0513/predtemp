```python
X_train = data.drop('A_temp', axis=1)
y_train = data['A_temp']
```

```python
TEST_PASS = 'data/2020.csv'

df = load_csv(TEST_PASS)
print(df.shape)
df.head()
```

```python
basename = os.path.splitext(os.path.basename(TEST_PASS))[0]
print(basename)
```

```python
X_test = df.drop('A_temp', axis=1)
y_test = df['A_temp']
```

```python
print(X_train.shape, X_test.shape)
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error


params = {
    'objective': 'regression',
    'metric': 'rmse',
    'boosting': 'gbdt',
    'learning_rate': 0.01,
    'max_depth': 7, # default -1 means no limit
    'num_leaves': 64, # default 31
    'min_child_samples': 20, # default 20
    'max_bin': 1000, # default 255 # 
    #'min_child_weight': 15, # default 1e-3 
    'feature_fraction': 0.9, # 
    'bagging_fraction': 0.75, # 
    'bagging_freq': 1, # 180
    'lambda_l1': 1e-06,
    'lambda_l2': 1e-06,
    'random_state': 42
   }

'''
params = {
    'objective': 'regression',
    'metric': 'rmse',
    'boosting': 'gbdt'
}
'''

y_preds = []
models = []
oof_train = np.zeros((len(X_train),))
list_rmse = []

kf = KFold(n_splits=4, shuffle=False)
kf_shuffle = KFold(n_splits=4, shuffle=True, random_state=71)
model_num = 0

for tr_index, val_index in kf.split(X_train):
    X_tr, X_val = X_train.iloc[tr_index], X_train.iloc[val_index]
    y_tr, y_val = y_train.iloc[tr_index], y_train.iloc[val_index]
    lgb_train = lgb.Dataset(X_tr, y_tr)
    lgb_eval = lgb.Dataset(X_val, y_val, reference=lgb_train)
    
    evals_results = {}
    model = lgb.train(
        params, lgb_train,
        valid_names=['train', 'valid'],
        valid_sets=[lgb_train, lgb_eval],
        evals_result=evals_results,
        verbose_eval=100,
        num_boost_round=10000,
        early_stopping_rounds=20
    )
    
    # 
    loss_train = evals_results['train']['rmse']
    loss_valid = evals_results['valid']['rmse']
    plt.plot(loss_train, label='train rmse')
    plt.plot(loss_valid, label='valid rmse')
    plt.ylabel('RMSE')
    plt.xlabel('Iterations')
    plt.title('best iteration is ' + str(model.best_iteration))
    plt.ylim()
    plt.xlim()
    plt.legend()
    plt.savefig('learning_curve_model' + str(model_num) + '.png', dpi=200)
    plt.show()

    y_tr_pred = model.predict(X_tr, num_iteration=model.best_iteration)
    oof_train[val_index] = model.predict(X_val, num_iteration=model.best_iteration)
    y_pred = model.predict(X_test, num_iteration=model.best_iteration)
    
    # RMSE
    df_rmse_cv = pd.DataFrame({'RMSE': [np.sqrt(mean_squared_error(y_tr, y_tr_pred))],
                               'RMSE': [np.sqrt(mean_squared_error(y_val, oof_train[val_index]))],
                               'RMSE': [np.sqrt(mean_squared_error(y_test, y_pred))]},
                             index=[model_num])
    df_rmse_cv.index.name = ''

    list_rmse.append(df_rmse_cv)
    df_rmse = pd.concat(list_rmse, axis=0)

    y_preds.append(y_pred)
    models.append(model)
    
    model_num += 1
```

```python
scores = [
    m.best_score['valid']['rmse'] for m in models
]

score = sum(scores) / len(scores)
print('===CV scores===')
print(scores)
print(score)
```
